```python
ratios = [
    'Profitability Ratios (Return from Sales) EBIT/Sales last avail. Year',
    'Profitability Ratios (Return from Sales) EBIT/Sales last year -1',
    'Profitability Ratios (Return from Sales) EBIT/Sales last year -2',
    'Profitability Ratios (Return on Investment) Net Profit/Total Assets last avail. Year',
    'Profitability Ratios (Return on Investment) Net Profit/Total Assets last year -1',
    'Profitability Ratios (Return on Investment) Net Profit/Total Assets last year -2',
    'Profitability Ratios (Return on Investment) Gross Profit/Total Assets last avail. Year',
    'Profitability Ratios (Return on Investment) Gross Profit/Total Assets last year -1',
    'Profitability Ratios (Return on Investment) Gross Profit/Total Assets last year -2',
    'Profit After Taxes/Turnover last available year',
    'Profit After Taxes/Turnover last year -1',
    'Profit After Taxes/Turnover last year -2',
    'Structure Ratios (Total-Assets-Structure Ratios) Current Assets/Total Assets last avail. Year',
    'Structure Ratios (Total-Assets-Structure Ratios) Current Assets/Total Assets last year -1',
    'Structure Ratios (Total-Assets-Structure Ratios) Current Assets/Total Assets last year -2',
    'Structure Ratios (Total-Assets-Structure Ratios) Inventory/Total Assets last avail. Year',
    'Structure Ratios (Total-Assets-Structure Ratios) Inventory/Total Assets last year -1',
    'Structure Ratios (Total-Assets-Structure Ratios) Inventory/Total Assets last year -2',
    'Structure Ratios (Total-Assets-Structure Ratios) Cash/Total Assets last avail. Year',
    'Structure Ratios (Total-Assets-Structure Ratios) Cash/Total Assets last year -1',
    'Structure Ratios (Total-Assets-Structure Ratios) Cash/Total Assets last year -2',
    'net working capital/total assets last avail. Year',
    'net working capital/total assets last year -1',
    'net working capital/total assets last year -2',
    'Cash Flow/Total Debt last available year',
    'Cash Flow/Total Debt last year -1',
    'Cash Flow/Total Debt last year -2',
    'Solvency Ratios Total Liabilities/Total Assets last avail. Year',
    'Solvency Ratios Total Liabilities/Total Assets last year -1',
    'Solvency Ratios Total Liabilities/Total Assets last year -2',
    'Trade Debts/Total Assets last available year',
    'Trade Debts/Total Assets last year -1',
    'Trade Debts/Total Assets last year -2',
    'Equity/Total Assets last available year',
    'Equity/Total Assets last year -1',
    'Equity/Total Assets last year -2',
    'Activity Ratios (Assets Turnover) Sales/total Fixed Assets last avail. Year',
    'Activity Ratios (Assets Turnover) Sales/total Fixed Assets last year -1',
    'Activity Ratios (Assets Turnover) Sales/total Fixed Assets last year -2',
    'Activity Ratios (Assets Turnover) Sales/Total Assets last avail. Year',
    'Activity Ratios (Assets Turnover) Sales/Total Assets last year -1',
    'Activity Ratios (Assets Turnover) Sales/Total Assets last year -2',
    'Long-Term Debt-to-Total-Assets Ratio last available year',
    'Long-Term Debt-to-Total-Assets Ratio last year -1',
    'Long-Term Debt-to-Total-Assets Ratio last year -2',
    'Liquidity Ratios (total liquid funds)/Current Liabilities last avail. Year',
    'Liquidity Ratios (total liquid funds)/Current Liabilities last year -1',
    'Liquidity Ratios (total liquid funds)/Current Liabilities last year -2',
    'Accounts Receivable/Total Assets last available year',
    'Accounts Receivable/Total Assets last year -1',
    'Accounts Receivable/Total Assets last year -2',
    'Liquidity Ratios (Current Liabilities − Cash)/Total Assets last avail. Year',
    'Liquidity Ratios (Current Liabilities − Cash)/Total Assets last year -1',
    'Liquidity Ratios (Current Liabilities − Cash)/Total Assets last year -2',
    'Quick Assets/Short-Term last available year',
    'Quick Assets/Short-Term last year -1',
    'Quick Assets/Short-Term last year -2',
    'Current Liabilities/Total Assets last available year',
    'Current Liabilities/Total Assets last year -1',
    'Current Liabilities/Total Assets last year -2'
]




ratiodf = pd.DataFrame(columns=ratios)
ratiodf.head()
```

#RATIO MAKING

```python
#Profitability Ratios
ratiodf['Profitability Ratios (Return from Sales) EBIT/Sales last avail. Year']= data['PROFIT/LOSS BEFORE TAXATIONth EURLast avail. yr'] / (data['Revenues from sales and servicesth EURLast avail. yr']+ 1e-9)
ratiodf['Profitability Ratios (Return from Sales) EBIT/Sales last year -1']= data['PROFIT/LOSS BEFORE TAXATIONth EURYear - 1'] / (data['Revenues from sales and servicesth EURYear - 1']+ 1e-9)
ratiodf['Profitability Ratios (Return from Sales) EBIT/Sales last year -2']= data['PROFIT/LOSS BEFORE TAXATIONth EURYear - 2'] / (data['Revenues from sales and servicesth EURYear - 2']+ 1e-9)
ratiodf['Profitability Ratios (Return on Investment) Net Profit/Total Assets last avail. Year']= data['PROFIT (LOSS)th EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Profitability Ratios (Return on Investment) Net Profit/Total Assets last year -1']= data['PROFIT (LOSS)th EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Profitability Ratios (Return on Investment) Net Profit/Total Assets last year -2']= data['PROFIT (LOSS)th EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Profitability Ratios (Return on Investment) Gross Profit/Total Assets last avail. Year']= data['Gross profitth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+ 1e-9)
ratiodf['Profitability Ratios (Return on Investment) Gross Profit/Total Assets last year -1'] = data['Gross profitth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Profitability Ratios (Return on Investment) Gross Profit/Total Assets last year -2']= data['Gross profitth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Profit After Taxes/Turnover last available year']= data['PROFIT (LOSS)th EURLast avail. yr']/(data['Revenues from sales and servicesth EURLast avail. yr']+1e-9)
ratiodf['Profit After Taxes/Turnover last year -1']= data['PROFIT (LOSS)th EURYear - 1']/(data['Revenues from sales and servicesth EURYear - 1']+1e-9)
ratiodf['Profit After Taxes/Turnover last year -2']= data['PROFIT (LOSS)th EURYear - 2']/(data['Revenues from sales and servicesth EURYear - 2']+1e-9)

#Structure Ratios
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Current Assets/Total Assets last avail. Year']= data['TOTAL CURRENT ASSETSth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Current Assets/Total Assets last year -1']= data['TOTAL CURRENT ASSETSth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Current Assets/Total Assets last year -2']= data['TOTAL CURRENT ASSETSth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Inventory/Total Assets last avail. Year']= data['TOTAL INVENTORIESth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Inventory/Total Assets last year -1']= data['TOTAL INVENTORIESth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Inventory/Total Assets last year -2']= data['TOTAL INVENTORIESth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Cash/Total Assets last avail. Year']= data['TOTAL LIQUID FUNDSth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Cash/Total Assets last year -1']= data['TOTAL LIQUID FUNDSth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Structure Ratios (Total-Assets-Structure Ratios) Cash/Total Assets last year -2']= data['TOTAL LIQUID FUNDSth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)

#Coverage Ratios
ratiodf['net working capital/total assets last avail. Year']= data['Net working capitalth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['net working capital/total assets last year -1']= data['Net working capitalth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['net working capital/total assets last year -2']= data['Net working capitalth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Cash Flow/Total Debt last available year']= (data['Gross profitth EURLast avail. yr']-data['TOTAL PRODUCTION COSTSth EURLast avail. yr'])/(data['TOTAL PAYABLESth EURLast avail. yr']+1e-9)
ratiodf['Cash Flow/Total Debt last year -1']= (data['Gross profitth EURYear - 1']-data['TOTAL PRODUCTION COSTSth EURYear - 1'])/(data['TOTAL PAYABLESth EURYear - 1']+1e-9)
ratiodf['Cash Flow/Total Debt last year -2']= (data['Gross profitth EURYear - 2']-data['TOTAL PRODUCTION COSTSth EURYear - 2'])/(data['TOTAL PAYABLESth EURYear - 2']+1e-9)

#Solvency
ratiodf['Solvency Ratios Total Liabilities/Total Assets last avail. Year']= data['TOTAL PAYABLESth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Solvency Ratios Total Liabilities/Total Assets last year -1']= data['TOTAL PAYABLESth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Solvency Ratios Total Liabilities/Total Assets last year -2']= data['TOTAL PAYABLESth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Trade Debts/Total Assets last available year']= data['Payables due within 12 monthsth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Trade Debts/Total Assets last year -1']= data['Payables due within 12 monthsth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Trade Debts/Total Assets last year -2']= data['Payables due within 12 monthsth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Equity/Total Assets last available year']= (data['TOTAL ASSETSth EURLast avail. yr']-data['TOTAL PAYABLESth EURLast avail. yr'])/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Equity/Total Assets last year -1']= (data['TOTAL ASSETSth EURYear - 1']-data['TOTAL PAYABLESth EURYear - 1'])/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Equity/Total Assets last year -2']= (data['TOTAL ASSETSth EURYear - 2']-data['TOTAL PAYABLESth EURYear - 2'])/(data['TOTAL ASSETSth EURYear - 2']+1e-9)

#Activity Ratios
ratiodf['Activity Ratios (Assets Turnover) Sales/total Fixed Assets last avail. Year']= data['Revenues from sales and servicesth EURLast avail. yr']/(data['TOTAL FIXED ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Activity Ratios (Assets Turnover) Sales/total Fixed Assets last year -1']= data['Revenues from sales and servicesth EURYear - 1']/(data['TOTAL FIXED ASSETSth EURYear - 1']+1e-9)
ratiodf['Activity Ratios (Assets Turnover) Sales/total Fixed Assets last year -2']= data['Revenues from sales and servicesth EURYear - 2']/(data['TOTAL FIXED ASSETSth EURYear - 2']+1e-9)
ratiodf['Activity Ratios (Assets Turnover) Sales/Total Assets last avail. Year']= data['Revenues from sales and servicesth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Activity Ratios (Assets Turnover) Sales/Total Assets last year -1']= data['Revenues from sales and servicesth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Activity Ratios (Assets Turnover) Sales/Total Assets last year -2']= data['Revenues from sales and servicesth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Long-Term Debt-to-Total-Assets Ratio last available year']= data['Payables due beyond 12 monthsth EURLast avail. yr']/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Long-Term Debt-to-Total-Assets Ratio last year -1']= data['Payables due beyond 12 monthsth EURYear - 1']/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Long-Term Debt-to-Total-Assets Ratio last year -2']= data['Payables due beyond 12 monthsth EURYear - 2']/(data['TOTAL ASSETSth EURYear - 2']+1e-9)

#Liquidity
ratiodf['Liquidity Ratios (total liquid funds)/Current Liabilities last avail. Year']= data['TOTAL LIQUID FUNDSth EURLast avail. yr']/(data['Payables due within 12 monthsth EURLast avail. yr']+1e-9)
ratiodf['Liquidity Ratios (total liquid funds)/Current Liabilities last year -1']= data['TOTAL LIQUID FUNDSth EURYear - 1']/(data['Payables due within 12 monthsth EURYear - 1']+1e-9)
ratiodf['Liquidity Ratios (total liquid funds)/Current Liabilities last year -2']= data['TOTAL LIQUID FUNDSth EURYear - 2']/(data['Payables due within 12 monthsth EURYear - 2']+1e-9)
ratiodf['Accounts Receivable/Total Assets last available year']= data['TOTAL RECEIVABLESth EURLast avail. yr']/data['TOTAL ASSETSth EURLast avail. yr']+1e-9
ratiodf['Accounts Receivable/Total Assets last year -1']= data['TOTAL RECEIVABLESth EURYear - 1']/data['TOTAL ASSETSth EURYear - 1']+1e-9
ratiodf['Accounts Receivable/Total Assets last year -2']= data['TOTAL RECEIVABLESth EURYear - 2']/data['TOTAL ASSETSth EURYear - 2']+1e-9
ratiodf['Liquidity Ratios (Current Liabilities − Cash)/Total Assets last avail. Year']= (data['Payables due within 12 monthsth EURLast avail. yr']-data['TOTAL LIQUID FUNDSth EURLast avail. yr'])/(data['TOTAL ASSETSth EURLast avail. yr']+1e-9)
ratiodf['Liquidity Ratios (Current Liabilities − Cash)/Total Assets last year -1']= (data['Payables due within 12 monthsth EURYear - 1']-data['TOTAL LIQUID FUNDSth EURYear - 1'])/(data['TOTAL ASSETSth EURYear - 1']+1e-9)
ratiodf['Liquidity Ratios (Current Liabilities − Cash)/Total Assets last year -2']= (data['Payables due within 12 monthsth EURYear - 2']-data['TOTAL LIQUID FUNDSth EURYear - 2'])/(data['TOTAL ASSETSth EURYear - 2']+1e-9)
ratiodf['Quick Assets/Short-Term last available year']= (data['TOTAL LIQUID FUNDSth EURLast avail. yr']+data['TOTAL RECEIVABLESth EURLast avail. yr'])/(data['Payables due within 12 monthsth EURLast avail. yr']+1e-9)
ratiodf['Quick Assets/Short-Term last year -1']= (data['TOTAL LIQUID FUNDSth EURYear - 1']+data['TOTAL RECEIVABLESth EURYear - 1'])/(data['Payables due within 12 monthsth EURYear - 1']+1e-9)
ratiodf['Quick Assets/Short-Term last year -2']= (data['TOTAL LIQUID FUNDSth EURYear - 2']+data['TOTAL RECEIVABLESth EURYear - 2'])/(data['Payables due within 12 monthsth EURYear - 2']+1e-9)
ratiodf['Current Liabilities/Total Assets last available year']= data['Payables due within 12 monthsth EURLast avail. yr']/data['TOTAL ASSETSth EURLast avail. yr']+1e-9
ratiodf['Current Liabilities/Total Assets last year -1']= data['Payables due within 12 monthsth EURYear - 1']/data['TOTAL ASSETSth EURYear - 1']+1e-9
ratiodf['Current Liabilities/Total Assets last year -2']= data['Payables due within 12 monthsth EURYear - 2']/data['TOTAL ASSETSth EURYear - 2']+1e-9
```

Threshold for extrem ratios
