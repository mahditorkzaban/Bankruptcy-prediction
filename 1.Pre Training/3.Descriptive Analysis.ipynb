```python


# Calculate the correlation matrix
correlation_matrix = data.corr()

# Display the correlation matrix as a heatmap
plt.figure(figsize=(40, 40))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix of Financial Ratios")
plt.show()
```

```python
correlation_matrix.to_excel('/content/drive/MyDrive/Bankruptcy Prediction Data/Visualization/Correlation.xlsx')
```

```python
# Extract and process the correlation matrix
correlation_matrix = data.corr()

# Flatten the correlation matrix and convert to DataFrame
correlation_pairs = correlation_matrix.unstack().reset_index()

# Rename columns for clarity
correlation_pairs.columns = ['Variable1', 'Variable2', 'Correlation']

# Remove self-correlations
correlation_pairs = correlation_pairs[correlation_pairs['Variable1'] != correlation_pairs['Variable2']]

# Drop duplicate pairs (e.g., keep (A, B) but drop (B, A))
correlation_pairs['AbsCorrelation'] = correlation_pairs['Correlation'].abs()
correlation_pairs = correlation_pairs.sort_values(by='AbsCorrelation', ascending=False).drop_duplicates(subset=['AbsCorrelation'])

# Filter pairs with high correlation (e.g., |correlation| > 0.8)
high_correlation_pairs = correlation_pairs[correlation_pairs['AbsCorrelation'] > 0.8]

# Save the highly correlated pairs to an Excel file
high_correlation_pairs.to_excel('/content/drive/MyDrive/Bankruptcy Prediction Data/Visualization/High_Correlation_Pairs.xlsx', index=False)

# Display the most correlated variable pairs
high_correlation_pairs
```
